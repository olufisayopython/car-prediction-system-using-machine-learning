```python
import pandas as pd
from sklearn import linear_model
df = pd.read_csv('car.csv')
df
```

```python
dummies = pd.get_dummies(df.Car)
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
final = merged.drop(['Car'],axis='columns')
final
```

```python
x = final.drop(['Price'],axis='columns')
x
```

```python
y = final.Price
y
```

```python
reg = linear_model.LinearRegression()
reg
```

```python
reg.fit(x,y)
```

```python
reg.predict([[67000,6,0,1,0]])
```

```python
reg.score(x,y)
```
